```python
import csv
from bs4 import BeautifulSoup
```

```python
def convert_html_to_csv(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as f:
        html_content = f.read()

    soup = BeautifulSoup(html_content, 'html.parser')

    # Find all the <a> tags containing video information
    video_entries = soup.find_all('div', {'class': 'mdl-grid'})

    # Prepare CSV writer
    with open(output_file, 'w', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow(['Video Title', 'Video URL', 'Watched Time'])

        for entry in video_entries:
            video_title = entry.find('a', {'class': 'pl-video-title-link'}).text.strip()
            video_url = entry.find('a', {'class': 'pl-video-title-link'})['href']
            watched_time = entry.find('span', {'class': 'pl-video-time'}).text.strip()

            writer.writerow([video_title, video_url, watched_time])

    print(f"Conversion completed. CSV file saved at: {output_file}")
```

```python
# Provide the file paths for input HTML and output CSV files
html_file = 'watch-history.html'
csv_file = 'converted-watch-history.csv'

# Convert HTML to CSV
convert_html_to_csv(html_file, csv_file)
```
